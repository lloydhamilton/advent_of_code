```python
import pandas as pd

sonar_readings = pd.read_csv('test_data.csv', header=None)
sonar_readings = sonar_readings.append(pd.Series(), ignore_index=True)
sonar_readings = sonar_readings.append(pd.Series(), ignore_index=True)
sonar_readings = sonar_readings.append(pd.Series(), ignore_index=True)
sonar_readings = sonar_readings.append(pd.Series(), ignore_index=True)
sonar_readings['lagged1'] = sonar_readings.iloc[:,0].shift(1)
sonar_readings['lagged2'] = sonar_readings.lagged1.shift(1)
sonar_readings['lagged3'] = sonar_readings.lagged2.shift(1)
sonar_readings['summed'] = sonar_readings.lagged1 + sonar_readings.lagged2 + sonar_readings.lagged3
sonar_readings['sum_lagged'] = sonar_readings.summed.shift(1)
sonar_readings['difference'] = (
    sonar_readings.summed > sonar_readings.sum_lagged 
)
sonar_readings
```

```python
sum(sonar_readings.difference)
```
