```python
import pandas as pd

data = pd.read_csv('test_data.csv', sep=' | ', engine='python', header=None)

def find_1478(data):
    output_values = data.iloc[:,-4:].copy()
    for x in output_values.index.values:
        output_values.iloc[x] = [len(out) for out in output_values.iloc[x].values]
    output_values = output_values.astype('int32')
    logic_df = (output_values == 4) | (output_values == 2) | (output_values == 3) | (output_values == 7)
    return logic_df.values.sum()

find_1478(data)
```

```python
# 9 minus 4 + 7 leaves 1
def return_diff(str1,str2):
    out = set(str1) - set(str2)
    if not out:
        return ""
    if out:
        return out

seven_string = row.loc[row.encoded == 7, 0].values[0]
four_string = row.loc[row.encoded == 4, 0].values[0]
joined_seven_4 = ''.join(set(''.join([seven_string, four_string])))
removed = [''.join(return_diff(char, joined_seven_4)) for char in row.loc[:,0].values]
row['nine'] = [len(x) for x in removed]
row.loc[(row.nine == 1) & (row.dist == 6),'encoded'] = 9
row
```
